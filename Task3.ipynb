```python
# 1: Employee Performance Bonus Eligibility
# Employee Performance Scores
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find highest score
highest_score = max(employees.values())

# Find employees who have the highest score (handle ties)
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

# Display result
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), 
      f"(Score: {highest_score})")
```

```python
# 2: Search Query Keyword Analysis
import string

query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Remove punctuation
query = query.translate(str.maketrans("", "", string.punctuation))

words = query.split()

# Count frequency
frequency = {}

for word in words:
    frequency[word] = frequency.get(word, 0) + 1

# Display words searched more than once
result = {}

for word, count in frequency.items():
    if count > 1:
        result[word] = count

print(result)
```

```python
# 3: Sensor Data Validation
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for hour, value in enumerate(sensor_readings):
    if value % 2 == 0:  # Even numbers are valid
        valid_readings.append((hour, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# 4: Email Domain Usage Analysis
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# Count domains
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

total_emails = len(emails)

# Calculate percentage
for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
# 5: Sales Spike Detection
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average
average_sales = sum(sales) / len(sales)

# Detect spikes (>30% above average)
threshold = average_sales * 1.30

for day, sale in enumerate(sales, start=1):
    if sale > threshold:
        print(f"Day {day}: {sale}")
```

```python
# 6: Duplicate User ID Detection
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

frequency = {}

# Count occurrences
for user in user_ids:
    frequency[user] = frequency.get(user, 0) + 1

# Display duplicates
for user, count in frequency.items():
    if count > 1:
        print(f"{user} → {count} times")
```
